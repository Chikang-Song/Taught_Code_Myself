**선택 사항: 게임스톱 주식 vs 테슬라**  
주식 가격을 결정하는 것은 복잡한 작업입니다. 이는 발행 주식 수, 회사의 미래 이익 규모 등 여러 요인에 따라 달라집니다. 중요한 요소 중 하나는 회사의 이익과 이익 성장입니다. 만약 회사의 이익이 증가하고 있다면, 주식 가격은 상승할 가능성이 높습니다. 회사 이익이 증가할 것으로 예상된다면, 주식을 매수하는 것이 좋습니다. 왜냐하면 주가가 상승할 가능성이 있기 때문입니다. 그러나 주가가 하락할 것으로 생각한다면 어떻게 해야 할까요?

주가 하락 시 수익을 얻는 방법은 **공매도(short selling)**입니다. 공매도는 투자자가 주식을 빌려 판매한 후, 주가가 하락했을 때 다시 매입하여 대여자에게 반환하는 방식입니다. 일반적으로 주가는 상승보다 하락이 더 빠르기 때문에 더 빨리 수익을 낼 수 있습니다. 대개 헤지펀드와 같은 숙련된 투자자들이 공매도에 참여합니다. 하지만 주가가 상승하면 투자자는 손실을 입을 수 있는 문제가 있습니다.

때로는 공매도 투자자들이 판단을 잘못 내리기도 합니다. 예를 들어 **테슬라**의 사례입니다. 몇 년 전, 많은 공매도 투자자들이 테슬라를 타겟으로 삼았습니다. 하지만 이후 테슬라가 수익을 내기 시작했고, 이익이 증가하면서 회사의 주가도 상승했습니다. 이러한 주가 상승은 회사의 실적에 기반한 것이기 때문에, 주가는 계속 상승할 가능성이 높습니다. 이와 달리, 최근 공매도된 주식이 기본적인 가치(fundamentals)와는 무관한 이유로 상승하는 경우도 있습니다. 이는 지속 가능성이 낮습니다.

**게임스톱**의 사례에서는 개인 투자자들이 레딧(Reddit) 온라인 커뮤니티인 **WallStreetBets**를 통해 게임스톱 주식을 대거 매수하기 시작했습니다. 게임 및 컴퓨터 게임 소매업체인 게임스톱은 당시 손실을 보고 있는 상황이었습니다. 그러나 수요의 급증으로 게임스톱의 주가는 폭등했습니다. 이로 인해 공매도를 진행했던 헤지펀드들에게는 수십억 달러의 손실이 발생했습니다. 게임스톱의 주가는 결국 하락할 가능성이 크기 때문에, 헤지펀드들은 공매도 포지션을 유지해야 합니다.  

헤지펀드에서 일하는 데이터 과학자로서, 테슬라와 게임스톱의 이익 데이터를 추출하고, 주식 가격과 이익을 비교하는 대시보드를 구축해야 합니다.
