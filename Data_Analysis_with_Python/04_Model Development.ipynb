### 📊 강의 전체 핵심 요약: 데이터 모델링과 평가의 A부터 Z까지

이 강의는 데이터 모델링의 핵심 개념부터 평가 방법까지 폭넓게 다룹니다. 아래는 주요 개념과 관련된 간단한 설명 및 예시를 포함한 요약입니다.

---

#### 1️⃣ **선형 회귀 (Linear Regression)**  
**핵심 내용**: 하나의 독립 변수를 사용하여 종속 변수의 값을 예측하는 방법입니다.  
- **수식**: $ y = b_0 + b_1x $
- **예시**: 자동차의 고속도로 연비(독립 변수)로 가격(종속 변수)을 예측.  
  예: 연비가 20이라면 가격은 $22,000.
  
- **독립 변수(Feature) vs. 종속 변수(Target Variable)**:  
  - **독립 변수 (X)**: 모델이 입력으로 받는 변수(예: `highway-mpg`).  
  - **종속 변수 (Y)**: 모델이 예측하려는 목표 값(예: `price`).  
  - `Y = df['price']`에서 **'price'가 종속 변수(Target Variable)** 임.  
  - `lm.fit(X, Y)`를 수행하면 모델이 `Y` 값을 예측하도록 학습됨.  

#### 2️⃣ **다중 선형 회귀 (Multiple Linear Regression)**  
**핵심 내용**: 여러 독립 변수를 사용하여 하나의 종속 변수를 설명하거나 예측합니다.  
- **수식**: $y = b_0 + b_1x_1 + b_2x_2 + \dots + b_nx_n$
- **예시**: 자동차의 연비, 연식, 제조사 등을 사용하여 자동차 가격을 예측.  
  예: $ \text{Price} = 50,000 - 2,000(\text{Age}) + 3,000(\text{Fuel Efficiency}) $


----
#### 3️⃣ **잔차 플롯과 모델 평가 (Residual Plots & Model Evaluation)**  
**핵심 내용**:  
- 잔차(residual)가 0을 중심으로 **무작위로 분포**한다면 선형 모델이 적절함.  
- **패턴이 보이면 선형 회귀 모델이 적합하지 않을 수 있음** → 비선형 모델(예: 다항 회귀) 고려 필요.  
- **좋은 선형 회귀 모델의 잔차 플롯 특징**:  
  - 잔차가 특정 패턴 없이 **무작위로 분포**.  
  - X축(예측값)에 대해 일정한 분산을 가짐(등분산성).  
  - 잔차가 점진적으로 증가하거나 감소하는 패턴을 보이지 않음.  
- **예시**:  
  - **적절한 모델**: 잔차가 0을 중심으로 무작위로 분포.  
  - **부적절한 모델**: 잔차가 곡선 패턴을 보이면 비선형 관계가 있을 가능성이 높음.  

---


----
#### 4️⃣ **다항 회귀 (Polynomial Regression)**  
**핵심 내용**: 선형 모델이 데이터에 적합하지 않을 경우 데이터를 다항식으로 변환하여 모델링.  
- **수식**: $ y = b_0 + b_1x + b_2x^2 + \dots $
- **예시**:  
  - 2차: $ y = -1.5x^2 + 5x + 3 $
  - 3차: $ y = 2x^3 - 4x^2 + 7x + 1 $  
- Python 활용: `sklearn.preprocessing.PolynomialFeatures`로 다항식 생성.  

- **R² 값과 비교할 때 주의할 점**:  
  - 일반적으로 **고차 다항 회귀 모델이 선형 회귀 모델보다 더 높은 R² 값을 가질 가능성이 큼**.  
  - 하지만 **R² 값만으로 모델의 성능을 비교하는 것은 적절하지 않음**.  
  - **고차 다항 회귀 모델은 과적합(Overfitting) 가능성이 크므로, 훈련 데이터에서는 높은 적합도를 보이지만 새로운 데이터에서는 성능이 저하될 수 있음**.  
  - 따라서 **MSE(Mean Squared Error), Cross-Validation 등 추가적인 평가 지표를 함께 고려해야 함**.  

---

#### 5️⃣ **평가 지표 (Evaluation Metrics)**  
모델의 성능을 평가하는 주요 지표:  

- **결정 계수 (R², Coefficient of Determination)**  
  - 모델이 데이터를 얼마나 잘 설명하는지를 나타내는 지표.  
  - 값 범위: **0 ≤ R² ≤ 1 (1에 가까울수록 적합도가 높음)**.  
  - **Python 코드에서 `lm.score(X, Y)`는 R² 값을 반환함**.  
  - R² 값이 **음수**라면, 모델이 평균값을 사용하는 것보다도 예측이 나쁨을 의미함.  

- **평균 제곱 오차 (MSE, Mean Squared Error)**  
  - 예측값과 실제값의 차이를 제곱한 후 평균을 구한 값.  
  - 값이 작을수록 모델의 예측이 더 정확함.  
  - **`mean_squared_error(y_actual, y_predicted)`를 사용하여 계산 가능**.  



---

#### 6️⃣ **데이터 파이프라인 (Data Pipelines)**  
**핵심 내용**: 데이터 변환 및 예측 과정을 간소화하기 위한 일련의 처리 흐름.  
- **예시**:  
  - 데이터 정규화 → 다항 변환 → 예측 순으로 파이프라인 작성.  
  - Python: `sklearn.pipeline.Pipeline`.  
- **장점**: 복잡한 과정을 하나의 객체로 처리 가능.




---

#### 7️⃣ **시각화를 활용한 모델 평가 (Visualization)**

<h2>1. Linear Regression and Multiple Linear Regression</h2>


단순 선형 회귀는 두 변수 간의 관계를 이해하는 데 도움을 주는 방법입니다:
* Variable X : 예측변수(Predictor) 혹은 독립변수 (Independent)
* Variable Y : 반응변수(Response) 혹은 종속변수 (Dependent)

<div class='alert alert-block alert-info'>선형 회귀의 결과는 선형 함수이며, 이는 예측 변수(독립 변수)를 기반으로 반응 변수(종속 변수)를 예측하는 함수</div>

 <b>Linear Function</b>
$$
\hat{Y} = a + b  X
$$


````python
# Load linear regression

from sklearn.linear_model import LinearRegression
````

Create the linear regression object:

````python
lm = LinearRegression()
lm
````

여기서는 단순 선형회귀를 사용해 고속도로 연비(highway-mpg)가 예측변수, 자동차 price가 종속 변수로 만듭니다.

````python
X = df[['highway-mpg']]
y = df['price']
````

<div class='alert alert-block alert-warning'>
<h4>✅ **1. scikit-learn의 입력 요구사항**</h4>
<div class='alert alert-block alert-info'>
    <ul>
        <li>
            <code>scikit-learn</code>의 <code>LinearRegression().fit(X, y)</code> 메서드는 입력 데이터를 다음과 같이 기대합니다:
        </li>
        <ol>
            <li>X (독립 변수, Feature Matrix)→2D 데이터프레임 (`pd.DataFrame`)</li>
            <li>y (종속 변수, Target Variable)→1D 시리즈 (`pd.Series`)</li>
        </ol>
    </ul>
    <p>
        <b>즉, X는 2차원 배열(행렬), y는 1차원 배열(벡터)여야 함.</b>
    </p>
</div>

* X를 **항상 2D 형식**(`DataFrame`)으로 유지하면 단일 변수뿐만 아니라 다중 변수도 처리 가능.
* 종속 변수는 **단일 목표 값(Scalar)으로 예측**해야 하므로 **1D 형식**(`Series`)이면 충분함.

만약 X가 1D `Series`라면 `scikit-learn`에서 예상하는 입력 형태와 다르기 때문에 오류가 발생할 수 있음.
</div>

고속도로 연비(highway-mpg)를 사용하여 선형 모델을 적합

예측 결과를 출력 합니다:

````python
Yhat = lm.predict(X)
Yhat[0:9]
````

`Intercept` (a)의 값:

````python
lm.intercept_
````

`slope` (b) 의 값:

````python
lm.coef_
````

<h4>Multiple Linear Regression</h4>

자동차 가격을 예측하기 위해 더 많은 변수를 모델에 추가하고 싶을때, **다중 선형 회귀(Multiple Linear Regression)**을 사용할 수 있습니다.

다중 선형 회귀는 단순 선형 회귀와 매우 유사하나, 하나의 연속적인 반응/종속 변수(y)와 **두 개 이상의** 예측/독립 변수(X) 간의 관계를 설명하는데 사용됩니다.

다음 예시는 네 개의 예측변수를 사용해 구조를 설명합니다.
$$
Y: Response \ Variable\\\\\\
X_1 :Predictor\ Variable \ 1\\
X_2: Predictor\ Variable \ 2\\
X_3: Predictor\ Variable \ 3\\
X_4: Predictor\ Variable \ 4\\
$$

$$
a: intercept\\\\\\
b_1 :coefficients \ of\ Variable \ 1\\
b_2: coefficients \ of\ Variable \ 2\\
b_3: coefficients \ of\ Variable \ 3\\
b_4: coefficients \ of\ Variable \ 4\\
$$

##### 주어진 수식:
$$
Yhat = a + b_1 X_1 + b_2 X_2 + b_3 X_3 + b_4 X_4
$$

````python
# 예측 변수 (Z = Xs)
Z = df[['horsepower','curb-weight','engine-size','highway-mpg']]
y = df['price']
````

````python
lm.fit(Z, y)
````

````python
lm.intercept_
````

````python
lm.coef_
````

이 예시로 얻을 수 있는 선형함수:

````python
print(f'Price = {lm.intercept_} + {lm.coef_[0]} x {Z.columns[0]} + {lm.coef_[1]} x {Z.columns[1]} + {lm.coef_[2]} x {Z.columns[2]} + {lm.coef_[-1]} x {Z.columns[-1]}')
````

<h3>(2) Residual Plot</h3>

데이터의 분산(variance)을 시각적으로 확인하는 좋은 방법 중 하나는 **잔차 그래프(residual plot)** 를 사용하는 것입니다.  

### **잔차(residual)란?**  
관측값(𝑦)과 예측값(ŷ, Yhat)의 차이를 **잔차(e)** 라고 합니다. 회귀 그래프에서 잔차는 데이터 포인트와 회귀선 사이의 거리입니다.  

### **잔차 그래프(residual plot)란?**  
잔차 그래프는 **잔차 값을 y축**, 독립 변수를 **x축** 에 나타낸 그래프입니다.  

### **잔차 그래프에서 주의해야 할 점**  
- **잔차가 x축을 기준으로 랜덤하게 분포되어 있다면**, **선형 모델이 적합** 하다는 의미입니다.  
  - 이유: 잔차가 랜덤하게 퍼져 있다는 것은 데이터의 분산이 일정하다는 뜻이며, 이는 선형 모델이 데이터를 잘 설명하고 있음을 나타냅니다.

````python
plt.figure(figsize=(12, 10))
sns.residplot(x = df['highway-mpg'], y=df['price']);
````

**이 잔차 그래프가 의미하는 바는 무엇인가요?**  

이 잔차 그래프를 보면, 잔차들이 x축을 중심으로 랜덤하게 분포하지 않고 일정한 패턴을 보이고 있습니다.  
이는 **선형 모델이 해당 데이터를 잘 설명하지 못할 가능성이 있으며, 비선형 모델이 더 적합할 수 있음** 을 시사합니다.

### **다중 선형 회귀 모델을 어떻게 시각화할 수 있을까요?**

<h2>3. Polynomial Regression and Pipelines</h2>


### **다항 회귀(Polynomial Regression)**  
다항 회귀는 **일반 선형 회귀 모델** 또는 **다중 선형 회귀 모델**의 특별한 경우입니다.  

- 예측 변수의 제곱을 사용하거나 **더 높은 차수의 항을 추가**하여 비선형 관계를 모델링할 수 있습니다.  

### **다항 회귀의 차수**  
다항 회귀는 차수(order)에 따라 다양한 형태를 가질 수 있습니다.  
차수가 증가할수록 모델은 데이터의 복잡한 패턴을 더 잘 포착할 수 있지만, 과적합(overfitting)의 위험도 커집니다.

<center><b>Quadratic - 2nd Order</b></center>
$$
Yhat = a + b_1 X +b_2 X^2 
$$


<center><b>Cubic - 3rd Order</b></center>
$$
Yhat = a + b_1 X +b_2 X^2 +b_3 X^3\\\\\\
$$


<center><b>Higher-Order</b>:</center>
$$
Y = a + b_1 X +b_2 X^2 +b_3 X^3 ....\\
$$


앞서 "highway-mpg"를 예측 변수로 사용할 때, 선형 모델이 최적의 적합도를 제공하지 않는다는 것을 확인했습니다.  

이제 **다항 회귀 모델(Polynomial Regression)** 을 사용하여 데이터를 더 잘 맞출 수 있는지 확인해 보겠습니다.

````python
def PlotPolly(model, independent_variable, dependent_variable, Name):
    x_new = np.linspace(independent_variable.min()-1, independent_variable.max()+1, 100)
    y_new = model(x_new)
    
    plt.plot(independent_variable, dependent_variable, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')
    
    plt.show()
    plt.close()
````

````python
# 변수들
x = df['highway-mpg']
y = df['price']
````

<div class='alert alert-block alert-warning'>
    <li>✔ <code>np.polyfit(x, y, 3)</code>는 x와 y가 1D(Series)여야 함</li>
    <li>✔ <code>np.polyfit()</code>은 numpy의 함수로, 단일 변수 다항 회귀만 수행 가능 → 다중 변수 지원 X</li>
    <li>✔ 다항 회귀에서는 1개의 독립 변수에 대해 다항식으로 변환하므로 Series(1D)로 전달 가능</li>
</div>

이제 **`polyfit`** 함수를 사용하여 다항 회귀 모델을 적합(fit)하고, **`poly1d`** 함수를 사용하여 다항식(polyynomial function)을 확인해 보겠습니다.

````python
# 여가서는 3차 다항식(큐빅, cubic)을 사용하여 회귀 모델을 적합합니다.
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
````

이제 함수를 그래프로 표현해 보겠습니다.

````python
PlotPolly(p, x, y, 'highway-mpg')
````

````python
np.polyfit(x, y, 3)
````

그래프를 보면, 이 다항 회귀 모델이 선형 모델보다 더 잘 작동하는 것을 확인할 수 있습니다.  
그 이유는 생성된 다항 함수가 더 많은 데이터 포인트를 잘 포착하고 있기 때문입니다.

<h2>Pipeline</h2>
데이터 파이프라인을 사용하면 데이터 처리 과정을 단순화할 수 있습니다.  

**`Pipeline`** 모듈을 사용하여 파이프라인을 생성하며, 데이터 정규화를 위해 **`StandardScaler`** 를 파이프라인의 한 단계로 추가할 수 있습니다.

````python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
````

파이프라인을 생성하려면 **모델(model) 또는 변환기(estimator)의 이름과 해당 생성자(Constructor)를 포함하는 튜플 리스트** 를 만듭니다.

````python
Input = [('scale',StandardScaler()), ('polynomial',PolynomialFeatures(include_bias=False)),
         ('model',LinearRegression())
        ]
````

생성된 튜플 리스트를 **파이프라인 생성자(Pipeline constructor)** 의 인수로 전달합니다.

````python
pipe = Pipeline(Input)
pipe
````

<div class='alert alert-block alert-info'>
    1️⃣ 데이터 표준화 (StandardScaler())
        <ul>
            <li>입력 데이터를 평균 0, 표준편차 1로 변환하여 모델이 학습하기 쉽게 만듦</li>
            <li>scale 단계에서 모든 특성(X 값)들이 동일한 스케일을 가지도록 조정함</li>
        </ul><br>
    2️⃣ 다항식 변환 (PolynomialFeatures(include_bias=False))
        <ul>
            <li>LinearRegression()이 원래 직선만 학습할 수 있으므로,→ 다항식 항(x², x³, ...)을 추가하여 곡선 회귀를 가능하게 만듦!</li>
            <li>include_bias=False: 상수항(1) 추가 X (LinearRegression이 자동으로 처리하므로 필요 없음)</li>
        </ul>
    3️⃣ 선형 회귀 모델 (LinearRegression())
    <ul>
        <li>이제 LinearRegression()이 직선이 아니라 변환된 다항식 데이터(X², X³ ...)를 학습하여 곡선 회귀를 수행함.</li>
    </ul>
</div>

먼저, **데이터 Z의 타입을 `float`로 변환** 하여, `StandardScaler`가 `float` 입력을 요구할 때 발생할 수 있는 변환 경고(conversion warning)를 방지합니다.  

그 후, 데이터를 **정규화(normalize)하고, 변환(transform)과 모델 적합(fit)을 동시에 수행** 할 수 있습니다.

````python
Z = Z.astype(float)
pipe.fit(Z, y)
````

마찬가지로, 데이터를 **정규화(normalize)하고, 변환(transform)한 후, 예측(prediction)을 동시에 수행** 할 수 있습니다.

<h2>4. Measures for In-Sample Evaluation</h2>


모델을 평가할 때는 **시각적 분석**뿐만 아니라, **정량적(quantitative) 평가**도 중요합니다.  

### **모델 정확도를 측정하는 두 가지 주요 지표**  
1. **R² (결정 계수, R-squared)**  

👉 R² 값은 회귀 모델이 실제 데이터를 얼마나 잘 설명하는지를 나타내는 척도입니다.

👉 즉, 예측값($\hat{y}$)이 실제값(𝑦)과 얼마나 가까운지를 측정합니다.

   - 회귀선이 데이터에 얼마나 잘 맞는지를 나타내는 지표입니다.  
   - R² 값은 **반응 변수(y)의 변동성 중에서 선형 모델이 설명하는 비율(%)** 을 의미합니다.  
   - 값이 1에 가까울수록 모델이 데이터를 잘 설명하고 있다는 뜻입니다.  

2. **평균 제곱 오차 (MSE, Mean Squared Error)**  
   - 모델의 예측값(ŷ)과 실제값(y) 간의 **오차를 제곱하여 평균을 낸 값** 입니다.  
   - 값이 작을수록 모델의 예측이 실제 데이터와 더 가깝다는 의미입니다.

# 📌 Task 3 - Polynomial Regression (다항회귀)

변수 **"CPU_frequency"** 를 사용하여 다항 특징(Polynomial Features) 을 생성합니다.  

✅ **3가지 다른 차수(degree)로 실험**  
- `numpy.polyfit`을 사용하여 **3가지 차수** (예: 2차, 3차, 4차 등)에서 다항 회귀를 수행하고
- 각 차수별 성능을 비교하여 **가장 적합한 차수를 찾습니다**. ✅  


````python
x = df['CPU_frequency']
y = df['Price']

#1d
f1 = np.polyfit(x, y, 1)
p1 = np.poly1d(f1)
print(p1)

# 3d
f3 = np.polyfit(x, y, 3)
p3 = np.poly1d(f3)
print(p3)

#5d
f5 = np.polyfit(x, y, 5)
p5 = np.poly1d(f5)
print(p5)
````

다음 단계로, **다항 회귀 결과를 실제 데이터 포인트와 비교하는 그래프** 를 그립니다. 

✅ **각 차수(degree)별 회귀 곡선과 실제 데이터 비교**  
- 다항 회귀 모델이 **실제 데이터에 얼마나 잘 맞는지 시각적으로 평가**(다음 `PlotPolly` 이용)
- 차수가 증가할수록 **과적합(overfitting)** 이 발생하는지도 확인 ✅  

````python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(independent_variable.min(),independent_variable.max(),100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title(f'Polynomial Fit for Price ~ {Name}')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
````

위 생성된 3개의 모델에 대해 그래프를 그립니다.

````python
PlotPolly(p1, x, y, 'CPU_frequency')
````

````python
PlotPolly(p3, x, y, 'CPU_frequency')
````

````python
PlotPolly(p5, x, y, 'CPU_frequency')
````

이제 각 **다항 회귀 모델의 성능을 평가** 하기 위해 **R² 값과 MSE 값을 계산** 합니다.

✅ **R² (결정 계수) 계산**

- 값이 **1에 가까울수록 모델이 데이터를 잘 설명함**  

✅ **MSE (평균 제곱 오차) 계산**  
- 값이 **작을수록 모델의 예측 오차가 적음**  

결과를 비교하여 **어떤 차수의 다항 회귀 모델이 가장 적합한지 평가** 합니다. ✅

# Task 4 - Pipeline

✅ 다음 단계를 포함하는 **파이프라인** 을 생성합니다.:  
1. **특성 스케일링 (Parameter Scaling)** → `StandardScaler()` 사용  
2. **다항 특성 생성 (Polynomial Feature Generation)** → `PolynomialFeatures()` 사용  
3. **선형 회귀 적용 (Linear Regression)** → `LinearRegression()` 사용  

✅ 이전과 동일한 **다중 특성(feature set)** 을 사용하여 파이프라인을 생성합니다.  
✅ 이 파이프라인을 활용하여 **모델을 학습(fit)하고 예측(predict) 수행** 합니다. 🚀

````python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
````

````python
Input = [
    ('scale',StandardScaler()),
    ('polynomial', PolynomialFeatures()),
    ('model', LinearRegression())
]
````

````python
pipe = Pipeline(Input)
````

````python
pipe
````

````python
Z = Z.astype(float)  # 에러방지 실수변환
pipe.fit(Z, y)
````

````python
ypipe = pipe.predict(Z)
ypipe[:10]
````

이제 **파이프라인을 통해 예측한 결과를 평가** 하기 위해 다음 값을 계산합니다:  

✅ **MSE (평균 제곱 오차)** → 값이 작을수록 예측 오차가 적음  
✅ **R² (결정 계수)** → 값이 1에 가까울수록 모델이 데이터를 잘 설명함  

결과를 분석하여 **이 파이프라인 모델이 기존 모델보다 성능이 좋은지 비교** 합니다. ✅🚀

````python
r2_score(y, ypipe)
````

````python
mean_squared_error(y, ypipe)
````

### **📊 R² 값의 변화 분석**  

✅ **단순 선형 회귀(SLR) → 다중 선형 회귀(MLR) → 다항 특성을 포함한 MLR**  
- **R² 값이 점점 증가** 하며 모델이 데이터를 더 잘 설명하게 됨  
- 다항 특성을 추가하면 **비선형 관계를 더 잘 포착** 하여 성능 향상  

### **📌 결론**  
**다중 선형 회귀(MLR) + 다항 특성** 을 적용하면 **가장 높은 R² 값** 을 얻을 수 있으므로,  
이 방법이 **가장 강력한 예측 모델** 이 될 가능성이 높습니다! ✅🚀

### ✅ **R² 값이 증가하는 이유: 단순 선형 회귀 → 다중 선형 회귀 → 다항 회귀**  

**질문:**  
👉 단순 선형 회귀 → 다중 선형 회귀 → 다항 회귀로 갈수록 R² 값이 증가하는 이유는?  
👉 다중 선형 회귀에 다항 특성을 추가하면 R² 값이 더 높아지는 이유는?  

✅ **단순 선형 회귀 → 다중 선형 회귀 → 다항 회귀로 갈수록 R² 값이 증가하는 이유**  
- **독립 변수를 추가하거나, 다항식 항을 추가하면 데이터의 변동성을 더 잘 설명할 수 있기 때문**  




👉 하지만, R² 값이 높아진다고 무조건 좋은 모델일까?  

✅ **하지만 R² 값이 높아진다고 무조건 좋은 모델이 아님**  
- **과적합(Overfitting) 가능성이 커질 수 있음**  
- **R² 외에도 MSE(평균 제곱 오차), Adjusted R² 등을 함께 평가해야 함**  



✅ **실제 머신러닝에서는 R² 값 하나만으로 모델을 평가하지 않고, 다양한 지표를 함께 고려해야 함** 🚀
