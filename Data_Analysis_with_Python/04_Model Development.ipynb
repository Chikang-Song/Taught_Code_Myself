<h2>3. Polynomial Regression and Pipelines</h2>


### **ë‹¤í•­ íšŒê·€(Polynomial Regression)**  
ë‹¤í•­ íšŒê·€ëŠ” **ì¼ë°˜ ì„ í˜• íšŒê·€ ëª¨ë¸** ë˜ëŠ” **ë‹¤ì¤‘ ì„ í˜• íšŒê·€ ëª¨ë¸**ì˜ íŠ¹ë³„í•œ ê²½ìš°ì…ë‹ˆë‹¤.  

- ì˜ˆì¸¡ ë³€ìˆ˜ì˜ ì œê³±ì„ ì‚¬ìš©í•˜ê±°ë‚˜ **ë” ë†’ì€ ì°¨ìˆ˜ì˜ í•­ì„ ì¶”ê°€**í•˜ì—¬ ë¹„ì„ í˜• ê´€ê³„ë¥¼ ëª¨ë¸ë§í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  

### **ë‹¤í•­ íšŒê·€ì˜ ì°¨ìˆ˜**  
ë‹¤í•­ íšŒê·€ëŠ” ì°¨ìˆ˜(order)ì— ë”°ë¼ ë‹¤ì–‘í•œ í˜•íƒœë¥¼ ê°€ì§ˆ ìˆ˜ ìˆìŠµë‹ˆë‹¤.  
ì°¨ìˆ˜ê°€ ì¦ê°€í• ìˆ˜ë¡ ëª¨ë¸ì€ ë°ì´í„°ì˜ ë³µì¡í•œ íŒ¨í„´ì„ ë” ì˜ í¬ì°©í•  ìˆ˜ ìˆì§€ë§Œ, ê³¼ì í•©(overfitting)ì˜ ìœ„í—˜ë„ ì»¤ì§‘ë‹ˆë‹¤.

<center><b>Quadratic - 2nd Order</b></center>
$$
Yhat = a + b_1 X +b_2 X^2 
$$


<center><b>Cubic - 3rd Order</b></center>
$$
Yhat = a + b_1 X +b_2 X^2 +b_3 X^3\\\\\\
$$


<center><b>Higher-Order</b>:</center>
$$
Y = a + b_1 X +b_2 X^2 +b_3 X^3 ....\\
$$


ì•ì„œ "highway-mpg"ë¥¼ ì˜ˆì¸¡ ë³€ìˆ˜ë¡œ ì‚¬ìš©í•  ë•Œ, ì„ í˜• ëª¨ë¸ì´ ìµœì ì˜ ì í•©ë„ë¥¼ ì œê³µí•˜ì§€ ì•ŠëŠ”ë‹¤ëŠ” ê²ƒì„ í™•ì¸í–ˆìŠµë‹ˆë‹¤.  

ì´ì œ **ë‹¤í•­ íšŒê·€ ëª¨ë¸(Polynomial Regression)** ì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ë” ì˜ ë§ì¶œ ìˆ˜ ìˆëŠ”ì§€ í™•ì¸í•´ ë³´ê² ìŠµë‹ˆë‹¤.

````python
def PlotPolly(model, independent_variable, dependent_variable, Name):
    x_new = np.linspace(independent_variable.min()-1, independent_variable.max()+1, 100)
    y_new = model(x_new)
    
    plt.plot(independent_variable, dependent_variable, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')
    
    plt.show()
    plt.close()
````

````python
# ë³€ìˆ˜ë“¤
x = df['highway-mpg']
y = df['price']
````

<div class='alert alert-block alert-warning'>
    <li>âœ” <code>np.polyfit(x, y, 3)</code>ëŠ” xì™€ yê°€ 1D(Series)ì—¬ì•¼ í•¨</li>
    <li>âœ” <code>np.polyfit()</code>ì€ numpyì˜ í•¨ìˆ˜ë¡œ, ë‹¨ì¼ ë³€ìˆ˜ ë‹¤í•­ íšŒê·€ë§Œ ìˆ˜í–‰ ê°€ëŠ¥ â†’ ë‹¤ì¤‘ ë³€ìˆ˜ ì§€ì› X</li>
    <li>âœ” ë‹¤í•­ íšŒê·€ì—ì„œëŠ” 1ê°œì˜ ë…ë¦½ ë³€ìˆ˜ì— ëŒ€í•´ ë‹¤í•­ì‹ìœ¼ë¡œ ë³€í™˜í•˜ë¯€ë¡œ Series(1D)ë¡œ ì „ë‹¬ ê°€ëŠ¥</li>
</div>

ì´ì œ **`polyfit`** í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‹¤í•­ íšŒê·€ ëª¨ë¸ì„ ì í•©(fit)í•˜ê³ , **`poly1d`** í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‹¤í•­ì‹(polyynomial function)ì„ í™•ì¸í•´ ë³´ê² ìŠµë‹ˆë‹¤.

````python
# ì—¬ê°€ì„œëŠ” 3ì°¨ ë‹¤í•­ì‹(íë¹…, cubic)ì„ ì‚¬ìš©í•˜ì—¬ íšŒê·€ ëª¨ë¸ì„ ì í•©í•©ë‹ˆë‹¤.
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
````

````python
PlotPolly(p, x, y, 'highway-mpg')
````

# ğŸ“Œ Task 1 : Single Linear Regression

ë°ì´í„°ì…‹ì˜ ì—¬ëŸ¬ íŠ¹ì„±(feature) ì¤‘ì—ì„œ **"CPU_frequency"** ê°€ ê°€ì¥ ë‚®ì€ p-valueë¥¼ ê°€ì§„ë‹¤ëŠ” ê²ƒì„ ì•Œì•˜ìŠµë‹ˆë‹¤.  

ì´ì œ **"CPU_frequency"** ì™€ **"Price"** ì˜ ê´€ê³„ë¥¼ ëª¨ë¸ë§í•˜ëŠ” **ë‹¨ì¼ íŠ¹ì„± ì„ í˜• íšŒê·€(SLR) ëª¨ë¸** ì„ ìƒì„±í•˜ì—¬ ê°€ê²©ì„ ì˜ˆì¸¡í•˜ëŠ” ëª¨ë¸ì„ ë§Œë“¤ì–´ ë³´ì„¸ìš”. âœ…

````python
# SLR model oject
lm = LinearRegression()

X = df[['CPU_frequency']]  # ë…ë¦½ë³€ìˆ˜
y = df['Price']  # ì¢…ì†ë³€ìˆ˜
````

````python
# ì˜ˆì¸¡ê²°ê³¼ Yhat
Yhat = lm.predict(X)
Yhat[:5]
````

ì˜ˆì¸¡ëœ ê°’ê³¼ ì‹¤ì œ ê°’ì˜ **ë¶„í¬ ê·¸ë˜í”„(Distribution Plot)** ë¥¼ ìƒì„±í•´ ì´ ê·¸ë˜í”„ë¥¼ í†µí•´ **ëª¨ë¸ì´ ì–¼ë§ˆë‚˜ ì˜ ì˜ˆì¸¡í–ˆëŠ”ì§€ í‰ê°€** í•©ë‹ˆë‹¤

ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•˜ê¸° ìœ„í•´ ë‹¤ìŒ ê°’ì„ ê³„ì‚°:

````python
mse = mean_squared_error(y, Yhat)  # MSE
mse
````

````python
r_sq = lm.score(X, y)  # R^2 ê²°ì •ê³„ìˆ˜
r_sq
````

# ğŸ“Œ Task 3 - Polynomial Regression (ë‹¤í•­íšŒê·€)

ë³€ìˆ˜ **"CPU_frequency"** ë¥¼ ì‚¬ìš©í•˜ì—¬ ë‹¤í•­ íŠ¹ì§•(Polynomial Features) ì„ ìƒì„±í•©ë‹ˆë‹¤.  

âœ… **3ê°€ì§€ ë‹¤ë¥¸ ì°¨ìˆ˜(degree)ë¡œ ì‹¤í—˜**  
- `numpy.polyfit`ì„ ì‚¬ìš©í•˜ì—¬ **3ê°€ì§€ ì°¨ìˆ˜** (ì˜ˆ: 2ì°¨, 3ì°¨, 4ì°¨ ë“±)ì—ì„œ ë‹¤í•­ íšŒê·€ë¥¼ ìˆ˜í–‰í•˜ê³ 
- ê° ì°¨ìˆ˜ë³„ ì„±ëŠ¥ì„ ë¹„êµí•˜ì—¬ **ê°€ì¥ ì í•©í•œ ì°¨ìˆ˜ë¥¼ ì°¾ìŠµë‹ˆë‹¤**. âœ…  


````python
x = df['CPU_frequency']
y = df['Price']

#1d
f1 = np.polyfit(x, y, 1)
p1 = np.poly1d(f1)
print(p1)

# 3d
f3 = np.polyfit(x, y, 3)
p3 = np.poly1d(f3)
print(p3)

#5d
f5 = np.polyfit(x, y, 5)
p5 = np.poly1d(f5)
print(p5)
````

ë‹¤ìŒ ë‹¨ê³„ë¡œ, **ë‹¤í•­ íšŒê·€ ê²°ê³¼ë¥¼ ì‹¤ì œ ë°ì´í„° í¬ì¸íŠ¸ì™€ ë¹„êµí•˜ëŠ” ê·¸ë˜í”„** ë¥¼ ê·¸ë¦½ë‹ˆë‹¤. 

âœ… **ê° ì°¨ìˆ˜(degree)ë³„ íšŒê·€ ê³¡ì„ ê³¼ ì‹¤ì œ ë°ì´í„° ë¹„êµ**  
- ë‹¤í•­ íšŒê·€ ëª¨ë¸ì´ **ì‹¤ì œ ë°ì´í„°ì— ì–¼ë§ˆë‚˜ ì˜ ë§ëŠ”ì§€ ì‹œê°ì ìœ¼ë¡œ í‰ê°€**(ë‹¤ìŒ `PlotPolly` ì´ìš©)
- ì°¨ìˆ˜ê°€ ì¦ê°€í• ìˆ˜ë¡ **ê³¼ì í•©(overfitting)** ì´ ë°œìƒí•˜ëŠ”ì§€ë„ í™•ì¸ âœ…  

````python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(independent_variable.min(),independent_variable.max(),100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title(f'Polynomial Fit for Price ~ {Name}')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
````

ìœ„ ìƒì„±ëœ 3ê°œì˜ ëª¨ë¸ì— ëŒ€í•´ ê·¸ë˜í”„ë¥¼ ê·¸ë¦½ë‹ˆë‹¤.

````python
PlotPolly(p1, x, y, 'CPU_frequency')
````

````python
PlotPolly(p3, x, y, 'CPU_frequency')
````

````python
PlotPolly(p5, x, y, 'CPU_frequency')
````

ì´ì œ ê° **ë‹¤í•­ íšŒê·€ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€** í•˜ê¸° ìœ„í•´ **RÂ² ê°’ê³¼ MSE ê°’ì„ ê³„ì‚°** í•©ë‹ˆë‹¤.

- ê°’ì´ **1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ëª¨ë¸ì´ ë°ì´í„°ë¥¼ ì˜ ì„¤ëª…í•¨**  

âœ… **MSE (í‰ê·  ì œê³± ì˜¤ì°¨) ê³„ì‚°**
