# Task 2: Overfitting
### 과적합 분석
데이터셋을 훈련 데이터와 테스트 데이터로 다시 분할하세요.

이번에는 데이터의 **50%**를 테스트용으로 예약합니다.
나머지 **50%**는 훈련 데이터로 사용합니다.

````python
# 데이터 분할 (50% 씩)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.5, random_state=0)
````

**"CPU_frequency"** 속성을 사용하여 다항 특성 (polynomial features)을 생성하세요.  
- 다항 차수를 **1부터 5까지** 변화시키면서 모델을 생성합니다.  
- 각 모델의 **R² 점수**를 평가합니다.  
- 결과값을 **리스트**에 저장하세요.

````python
lre = LinearRegression()
Rsqu_test = []
order = [1, 2, 3, 4, 5]
for n in order:
    pr = PolynomialFeatures(degree=n)
    x_train_pr = pr.fit_transform(x_train[['CPU_frequency']])
    x_test_pr = pr.fit_transform(x_test[['CPU_frequency']])    
    lre.fit(x_train_pr, y_train)
    Rsqu_test.append(lre.score(x_test_pr, y_test))
````

# Task 3 : Ridge Regression

다음 여러 개의 특성(feature) 을 사용하여 다항 특성 모델을 생성하세요:

'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category'
해야 할 작업:

차수(degree) = 2인 다항 특성 모델 생성
훈련 데이터 및 테스트 데이터 속성 집합 생성

````python
# Write your code below and press Shift+Enter to execute 
pr=PolynomialFeatures(degree=2)
x_train_pr=pr.fit_transform(x_train[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']])
x_test_pr=pr.fit_transform(x_test[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']])
````

**Ridge 회귀 모델**을 생성하고 평가하세요.  

- **하이퍼파라미터 α(alpha)** 값을 **0.001에서 1까지**, **0.001 단위로 증가**시키며 테스트합니다.  
- 각 **α 값에 대한 R² 점수**를 훈련 데이터와 테스트 데이터에서 계산합니다.  
- **훈련 데이터 R² 점수 리스트**와 **테스트 데이터 R² 점수 리스트**를 생성하세요.

````python
Rsqu_test = []
Rsqu_train = []
Alpha = np.arange(0.001,1,0.001)
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha) 
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})
    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
````

훈련 데이터와 테스트 데이터의 **R² 값**을 **α(alpha) 값**에 따라 그래프로 나타내세요.  

- **X축**: α 값 (0.001 ~ 1)  
- **Y축**: R² 점수  
- 훈련 데이터와 테스트 데이터를 **각각 다른 색상**으로 표시하세요.

````python
plt.figure(figsize=(10, 6))  
plt.plot(Alpha, Rsqu_test, label='validation data')
plt.plot(Alpha, Rsqu_train, 'r', label='training Data')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.ylim(0, 1)
plt.legend()
````

- **GridSearchCV**를 사용하여 최적의 **α 값**을 찾으세요.  
- 모델이 가장 좋은 성능을 보이는 **최적의 α 값**을 출력하세요.  


````python
parameters1= [{'alpha': [0.0001,0.001,0.01, 0.1, 1, 10]}]
````

**Ridge 회귀 인스턴스**를 생성하고 **4-폴드 교차 검증**을 사용하여 **Grid Search**를 실행하세요.  

- **Ridge 회귀 모델** 생성  
- **GridSearchCV**를 활용하여 최적의 **α 값** 탐색  
- **4-폴드 교차 검증** 적용

````python
RR=Ridge()
Grid1 = GridSearchCV(RR, parameters1,cv=4)
````

Grid Search를 **훈련 데이터**에 적합(fit)시키세요.  
- 최적의 하이퍼파라미터 **α 값**을 찾기 위해 **GridSearchCV**를 훈련 데이터에 적용합니다.

````python
Grid1.fit(x_train[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core', 'OS', 'GPU', 'Category']], y_train)
````

Grid Search를 통해 찾은 **최적의 α 값**을 사용한 모델의 **테스트 데이터 R² 점수**를 출력하세요.

````python
BestRR=Grid1.best_estimator_
print(BestRR.score(x_test[['CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'CPU_core','OS','GPU','Category']], y_test))
````

## 📌 **Python 데이터 분석 치트시트: 모델 평가 및 개선**  

### **1. 훈련 데이터와 테스트 데이터 분할**  
```python
from sklearn.model_selection import train_test_split

y_data = df['target_attribute']
x_data = df.drop('target_attribute', axis=1)

x_train, x_test, y_train, y_test = train_test_split(
    x_data, y_data, test_size=0.10, random_state=1
)
```
🔹 **설명:**  
- `target_attribute`(예측할 목표 변수)를 분리  
- 나머지 데이터를 입력 변수로 사용  
- `train_test_split()`을 사용하여 **10%를 테스트 데이터**로 분할  

---

### **2. 교차 검증 (Cross Validation Score)**  
```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression 

lre = LinearRegression()
Rcross = cross_val_score(lre, x_data[['attribute_1']], y_data, cv=n)  
# n은 교차 검증을 수행할 폴드(folds) 개수

Mean = Rcross.mean()  # 평균 R² 값
Std_dev = Rcross.std()  # 표준 편차
```
🔹 **설명:**  
- 데이터가 충분하지 않을 때, **교차 검증**을 사용하여 여러 번 모델을 평가  
- `cv=n`을 설정해 **n개 폴드로 나누어 검증**  
- 평균(`mean()`)과 표준 편차(`std()`)를 계산하여 모델 성능 확인  

---

### **3. 교차 검증을 활용한 예측 (Cross Validation Prediction)**  
```python
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LinearRegression 

lre = LinearRegression()
yhat = cross_val_predict(lre, x_data[['attribute_1']], y_data, cv=4)
```
🔹 **설명:**  
- **교차 검증을 수행하면서 예측값(`yhat`)을 생성**  
- 데이터가 부족할 때 모델 성능을 더 안정적으로 평가 가능  

---

### **4. Ridge Regression (릿지 회귀) 및 예측**  
```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures

pr = PolynomialFeatures(degree=2)  
x_train_pr = pr.fit_transform(x_train[['attribute_1', 'attribute_2', ...]])
x_test_pr = pr.transform(x_test[['attribute_1', 'attribute_2', ...]])

RidgeModel = Ridge(alpha=1)
RidgeModel.fit(x_train_pr, y_train)
yhat = RidgeModel.predict(x_test_pr)
```
🔹 **설명:**  
- **다항 회귀 모델을 개선하기 위해 Ridge Regression 적용**  
- `alpha` 값을 설정하여 **과적합(overfitting) 방지**  
- Ridge는 **고차원의 특성이 모델에 미치는 영향을 조절**  

---

### **5. Grid Search (최적의 하이퍼파라미터 찾기)**  
```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

parameters = [{'alpha': [0.001, 0.1, 1, 10, 100, 1000, 10000]}]
RR = Ridge()
Grid1 = GridSearchCV(RR, parameters, cv=4) 
Grid1.fit(x_data[['attribute_1', 'attribute_2', ...]], y_data)

BestRR = Grid1.best_estimator_
BestRR.score(x_test[['attribute_1', 'attribute_2', ...]], y_test)
```
🔹 **설명:**  
- **Grid Search를 활용해 최적의 `alpha` 값 찾기**  
- **교차 검증(cv=4) 사용하여 최적의 모델을 평가**  
- `Grid1.best_estimator_`를 사용하여 **최적 모델 선택**  
- 최종적으로 **테스트 데이터에서 모델의 성능(R² 값) 평가**
